# Capture

```julia
push!(LOAD_PATH,"../src/")
using CaptureRate
using Plots
using CSV
```

```julia
# import data from two files
# Q (first column, amu) and E (second column, eV)
pot_1 = CSV.read("../notebook/Potential_1.csv")
pot_2 = CSV.read("../notebook/Potential_2.csv")
potential_matrix_1 = convert(Matrix, pot_1)
potential_matrix_2 = convert(Matrix, pot_2)
poly_order = 4;
```

```julia
cc = CaptureRate.calc_poly_wave_func(potential_matrix_1, potential_matrix_2, poly_order, -5, 10, 500, 25, 50);
CaptureRate.calc_overlap!(cc; cut_off=0.05, σ=0.01)
```

```julia
CaptureRate.plot_potentials(cc, scale_factor=0.01)
ylims!(0, 0.5)
xlims!(-4, 8)
```
