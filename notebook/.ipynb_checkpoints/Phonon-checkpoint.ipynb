# Phonon

```julia
push!(LOAD_PATH,"../src/")
using Phonon
using Plots
# using Brooglie
#plotly() # Choose the Plotly.jl backend for web interactivity
```

```julia
ħω = 0.04
NQ = 1000
Qi = -10
Qf = 10
x =linspace(Qi, Qf, NQ)
```

```julia
plot()

e, v = Phonon.solve1D_ev_amu(x->Phonon.harmonic(x, ħω)-1, NQ=NQ , Qi=Qi, Qf=Qf, m=1, nev=50)
# print(e)
for i = 1:length(e)
    plot!(x, v[i]*0.2+e[i], color="#4575b4")
end

plot!(x, Phonon.harmonic(x, ħω)-1, lw=2, color="black")
ylims!(-1.5, 2.2)
xlims!(-2, 4.2)
```

```julia
ħω = 0.04
NQ = 1000
Qi = -5
Qf = 5
x =linspace(Qi, Qf, NQ)
plot()
# f(x) = 0.1*ħω*x.*x.*x.*x - 2*ħω*x.*x 
# f(x) = 0.5*ħω^2*x.*x

# f(x) = Phonon.double(x, 1*ħω*4, 0.2*ħω*15)
f(x) = Phonon.step(x, 2,2)

e, v = Phonon.solve1D_ev_amu(f, NQ=NQ, Qi=Qi, Qf=Qf, m=1, nev=22, maxiter=1000)
# print(e)
plot()
for i = 1:length(e)
    plot!(x, v[i]*0.2+e[i], color="#4575b4")
end

plot!(x, f(x), lw=2, color="black")
ylims!(-5.5, 2.2)
xlims!(-5, 5)
```
