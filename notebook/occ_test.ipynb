```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T = np.arange(0.1, 100, 0.1)
```

```python
omega = 1

for n in range(121):
    Z = 1/(1-np.exp(-omega/T))
    w = np.exp(-n*omega/T) / Z

    plt.plot(T, w)
```

```python
ns = []
for t in T:
    exp_n = 0
    Z = 1/(1-np.exp(-omega/t))
    for n in range(501):
        w = np.exp(-n*omega/t) / Z
        exp_n += n*w
    ns.append(exp_n)

plt.plot(T, ns)
```
